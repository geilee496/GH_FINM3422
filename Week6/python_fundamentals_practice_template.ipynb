## 6) Control Flow: if / elif / else

```python
# 6) Grading rule:
# score >= 85 -> "HD"
# score >= 75 -> "D"
# score >= 65 -> "C"
# score >= 50 -> "P"
# else        -> "N"
# Implement function grade(score) returning the string grade.

def grade(score):
    # YOUR CODE HERE
    if score>=85:
        return "HD" #if score > than 85, u get HD
    elif score >=75: #else if score is not > than 85 but >/= to 75 u get D
        return "D"
    elif score >=65:
        return "C"
    elif score >=50:
        return "P"
    else:
        return "N"
    return None


# CHECKS
assert grade(90) == "HD"
assert grade(78) == "D"
assert grade(66) == "C"
assert grade(52) == "P"
assert grade(42) == "N"
```

## 7) Loops: for/while, enumerate/range

```python
# 7A) Sum of squares from 1..N (inclusive):
# Implement sum_of_squares(N) using a for-loop.

def sum_of_squares(N):
    total = 0 #i think 0 is 1
    for number in range(1,N+1):
        total += number ** 2 #returns current total + what im addin which is number squared
    return total #if theres no return rn so it'll return with None but we want total, 
# we start bu adding to 1 to total, then squaring it and summing them togethr ig
    return None

# CHECK
assert sum_of_squares(3) == 1**2 + 2**2 + 3**2
```

```python
# 7B) Re-implement sum_of_squares using a generator expression (one line).
def sum_of_squares_fast(N):
    return sum(i ** 2 for i in range(N+1))
    return None

# CHECK
assert sum_of_squares_fast(5) == 55
```

```python
# 8B) Given a list of monthly returns (e.g., [0.01, -0.02, 0.015]),
# compute a wealth index starting at 1.0 and return the final value.
# Example: wealth([0.01, -0.02]) -> 1.0 * 1.01 * 0.98 = 0.9898

def wealth(returns):
    w = 1 # w is wealth, starts wealth at 1
    for r in returns:
        w *= (1+r) #for each returns, u multiply ur wealth through and then return the wealth
    return None

# CHECK
assert abs(wealth([0.01, -0.02]) - (1.0*1.01*0.98)) < 1e-12
```

## 9) Mini‑Project (bonus): Tiny expense parser (strings → dict)

```python
#not necessary to know, wants us to return a dictionary
data = """
groceries,12.50
transport,4.20
groceries,17.30
coffee,3.80
"""

# Build a dict totals so that:
# totals["groceries"] == 29.8
# totals["transport"] == 4.2
# totals["coffee"] == 3.8

totals = {}
for line in data.strip().splitlines(): #splitting all new line characters and replacing all of them?
    name, value = line.split(",")
    value = float(value) #so covert value to a float
    totals[name] = totals.get(name,0) + value



# CHECK
assert totals == {"groceries": 29.8, "transport": 4.2, "coffee": 3.8}
```

```python
# Write the totals dict (from section 9) to a file 'results.txt' as "key,value" each line.
# Then read it back and confirm the contents match.

#last yr this was somewhat a part of our assignment

# YOUR CODE HERE
with open("results.txt", "w", encoding = "utf-8") as file:
    for k,v in totals.items(): #using total dictionary from above in Q9
        file.write(f"{k},{v}\n")


# CHECK (lightweight)
with open('results.txt','r',encoding='utf-8') as f:
    content = f.read().strip().splitlines()
assert 'groceries,29.8' in content
```
